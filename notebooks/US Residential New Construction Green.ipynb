## Typical RECS Charts and Findings on Energy Efficiency Measures:

### 1. **Insulation Levels by Climate Zone**
```
Chart Type: Bar chart
Shows: Percentage of homes with recommended insulation levels
Pattern: Higher insulation adoption in colder climate zones (5-7) vs warmer zones (1-3)
Data: Typically shows 60-80% compliance in northern zones vs 30-50% in southern zones
```

### 2. **Window Efficiency by Region**
```
Chart Type: Stacked bar chart
Shows: Single-pane vs double-pane vs ENERGY STAR window adoption
Trend: Significant increase in efficient windows in new construction (80%+ double-pane)
```

### 3. **Heating System Efficiency by Vintage**
```
Chart Type: Line chart over time
Shows: AFUE ratings of heating systems in new vs existing homes
Finding: New construction averages 90%+ AFUE vs 60-70% for older homes
```

### 4. **Energy Star Appliance Penetration**
```
Chart Type: Pie charts by appliance type
Data: Refrigerators (45%), Clothes washers (40%), Dishwashers (35%) in new homes
```
